```python
import json
from pathlib import Path

from ai_scoring import score_apartment
```

## 2. Укажи путь к фото

Вставь путь к своему изображению комнаты (jpg/png).

```python
IMAGE_PATH = "room.jpg"  # <- замени на путь к своему фото
```

```python
# Показать фото (если есть)
from PIL import Image

if Path(IMAGE_PATH).exists():
    img = Image.open(IMAGE_PATH)
    display(img)
else:
    print(f"Файл не найден: {IMAGE_PATH}")
    print("Укажи правильный путь в ячейке выше.")
```

## 3. Запуск оценки

```python
result = score_apartment(IMAGE_PATH)
```

## 4. Результат (JSON)

```python
print(json.dumps(result, ensure_ascii=False, indent=2))
```

```python
# Краткая сводка
print(result.get("summary", ""))
print()
print("Оценки:")
for k, v in result.items():
    if k != "summary" and isinstance(v, (int, float)):
        print(f"  {k}: {v}/10")
```
