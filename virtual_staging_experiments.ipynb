# Задание 2: Эксперименты с гиперпараметрами Virtual Staging (GPT Image 1.5)

Документация влияния параметров на итоговую картинку.

## Гиперпараметры GPT Image 1.5

| Параметр | Значения | Влияние |
|----------|----------|----------|
| **input_fidelity** | low, high | low = сильнее изменения, больше творчества. high = сохраняет детали исходника |
| **quality** | low, medium, high, auto | Выше = лучше качество, дольше. low = быстрее и дешевле |
| **size** | auto, 1024x1024, 1536x1024, 1024x1536 | auto = пропорции исходника. Фикс = обрезка/растяжка |

```python
from dotenv import load_dotenv
load_dotenv()

from virtual_staging import virtual_stage, StagingConfig

IMAGE_PATH = "room.jpg"  # <- твоё фото
PROMPT = "Replace the old sofa with a modern gray scandinavian sofa"
```

## Тест 1: input_fidelity

- **low** — модель сильнее меняет сцену, лучше для замены мебели
- **high** — сохраняет стиль и детали исходного фото

```python
config_low = StagingConfig(input_fidelity="low", quality="high", size="auto")
img_low = virtual_stage(IMAGE_PATH, PROMPT, "staged_fidelity_low.jpg", config=config_low)
display(img_low)
```

```python
config_high = StagingConfig(input_fidelity="high", quality="high", size="auto")
img_high = virtual_stage(IMAGE_PATH, PROMPT, "staged_fidelity_high.jpg", config=config_high)
display(img_high)
```

## Тест 2: quality

- **low** — быстрее, может быть размытие
- **high** — максимум деталей

```python
config_qlow = StagingConfig(quality="low")
virtual_stage(IMAGE_PATH, PROMPT, "staged_quality_low.jpg", config=config_qlow)
```

```python
config_qhigh = StagingConfig(quality="high")
virtual_stage(IMAGE_PATH, PROMPT, "staged_quality_high.jpg", config=config_qhigh)
```
